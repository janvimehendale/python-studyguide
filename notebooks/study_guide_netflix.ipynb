```python
# Convert a column that has some non-numeric values
# This will turn non-numeric values into NaN (missing values)
netflix['duration_numeric'] = pd.to_numeric(netflix['duration'], errors='coerce')
#The  errors='coerce'  parameter means: "If you can't convert it to a number, turn it into NaN (missing value) instead of crashing."
netflix
```

```python
# See what type each column is
netflix.dtypes
```

```python
# Check if a specific column is numeric
netflix['duration_numeric'].dtypes
```

```python
# Count how many times each country appears
country_count = netflix['country'].value_counts()
country_count
```

```python
# Get the most frequent (first one)
most_common = country_count.index[0]
print(f'Most Common Country : {most_common}') 
```

```python
most_common = netflix['country'].value_counts().index[0]
most_common
```

```python
# Theduration column is formatted like "90 min"
# Split by the space to get time length
# Method: Use str.split()
netflix[['time', 'length']] = netflix['duration'].str.split(' ', expand=True)
netflix
```

```python
# If you want to extract parts without a delimiter:
netflix['month_added'] = netflix['date_added'].str[:9]
netflix['date_year'] = netflix['date_added'].str[10:]
netflix
```

```python
netflix['rating'].unique()
```

```python
# Create a mapping dictionary
rating_mapping = {
    "G" : "General Audiences",
    "PG" : "Parental Guidance",
    "R" : "Restricted"
}
```

```python
# Replace the values
netflix['rating_full'] = netflix['rating'].map(rating_mapping)
netflix
```

```python
# Check the first few values
netflix[['title','director','cast']].head()
```

```python
# Find the row with the highest year
max_year = netflix['release_year'].idxmax()
max_year
```

```python
max_movie_year = netflix.loc[max_year, 'release_year']
max_movie_year
```

```python
#Filter the year after 2020
netflix_after_2020 = netflix[netflix['release_year']> 2020]
netflix_after_2020
```

```python
# Filter for a specific genre
netflix_documnetaries = netflix[netflix['listed_in']== 'Documentaries']
netflix_documnetaries
```

```python
# Get documentaries and release year post 2020 both
filtered = netflix[(netflix['listed_in']== 'Documentaries') & (netflix['release_year']> 2020)]
filtered
```

```python
# Get documentaries OR release year post 2020 
filtered_or = netflix[(netflix['listed_in']== 'Documentaries') | (netflix['release_year']> 2020)]
filtered_or
```

```python
# For each director, how many different genres have they worked in?
# Group by director and count unique genres
genres_per_director = netflix.groupby('director')['listed_in'].nunique()
genres_per_director
```

```python
# Find the director with the most genres
versatile_director = genres_per_director.idxmax()
genre_count = genres_per_director.max()
print(f"{versatile_director} has worked in {genre_count} different genres")
```

```python
# Count how many times each actor appears as a lead actor
cast_counts = netflix['cast'].value_counts()
cast_counts
```

```python
#Bar Charts (Counting Categories)
#Shows how many times each category appears.
# Count how many movies are in each genre
genre_counts = netflix['listed_in'].value_counts()
# Create a bar chart of the top 5
genre_counts.head(5).plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Top 5 Movie Genres')
plt.show()
```
